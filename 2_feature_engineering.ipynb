```python
# 数值特征
import pandas as pd
import numpy as np
import json
from tqdm import tqdm as tqdm
from sklearn.model_selection import KFold
```

```python
train_data = []
train_data_dict = {}
with open('data/raw_data/train.json', 'r') as f:
    for line in f:
        raw = eval(str(json.loads(line)).lower())
        train_data.append(raw)
        #train_data_dict[raw['text_id']] = raw
kb = {}
with open('data/raw_data/kb_data', 'r') as f:
    for line in f:
        item = eval(str(json.loads(line)).lower())
        kb[item['subject_id']] = item
```

```python
data = pd.read_pickle('features/1_stat_feature.pkl')
```

```python
# rank 特征
to_rank = [
           'num_candidates','mention_equal_subject','shuminghao','entity_common',
           'cos_distance','mse_distance',
           'numerical_0', 'numerical_1', 'numerical_2',
       'numerical_3', 'numerical_4', 'numerical_5', 'numerical_6',
       'numerical_7', 'numerical_8', 'numerical_9', 'numerical_10',
       'numerical_11', 'numerical_12', 'numerical_13', 'numerical_14'
          ]
for f in to_rank:
    data_group = data.groupby('kb_id')[f].rank(ascending=False)
    data['kbid_rank_'+f] = data_group
    try:
        data['kbid_rank_'+f] = data['kbid_rank_'+f].astype('int32')
    except:
        pass
```

```python
# mean and var
to_rank = [
           'num_candidates','mention_equal_subject','shuminghao','entity_common',
           'cos_distance','mse_distance',
           'numerical_0', 'numerical_1', 'numerical_2',
       'numerical_3', 'numerical_4', 'numerical_5', 'numerical_6',
       'numerical_7', 'numerical_8', 'numerical_9', 'numerical_10',
       'numerical_11', 'numerical_12', 'numerical_13', 'numerical_14'
          ]
for f in tqdm(to_rank):
    data_group = data.groupby('kb_id')[f].agg({'kb_'+f+'_var':'var','kb_'+f+'_mean':'mean','kb_'+f+'_max':'max',
                                                   'kb_'+f+'_min':'min'})
    data = data.merge(data_group,on='kb_id',how='left')
```
