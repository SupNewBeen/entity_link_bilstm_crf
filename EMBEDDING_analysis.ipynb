```python
# coding:u8
import numpy as np
from utils import seed_torch, read_data, load_glove, get_threshold
from tokenize_pkg.tokenize import Tokenizer
```

```python
embedding_baike = '/home/zhukaihua/Desktop/nlp/embedding/baike'
embedding_qq = '/home/zhukaihua/Desktop/nlp/embedding/Tencent_AILab_ChineseEmbedding.txt'
```

```python
def get_coefs(word, *arr):
    return word, np.asarray(arr, dtype='float32')

embeddings_baike = dict(get_coefs(*o.rstrip(' \n').split(" ")) for o in open(embedding_baike) if len(o) > 100)
embeddings_qq = dict(get_coefs(*o.rstrip(' \n').split(" ")) for o in open(embedding_qq) if len(o) > 100)
```

```python
train_X, train_label, dev_X, dev_label = read_data()
assert len(train_X) == len(train_label)
train_X = [' '.join(word) for word in train_X]
dev_X = [' '.join(word) for word in dev_X]
t = Tokenizer(max_feature=500000, segment=False)
t.fit(list(train_X) + list(dev_X))
```

```python
baike_notfound = []
qq_notfound = []
for item in t.voc.keys():
    if item not in embeddings_baike:
        for sub in jieba.lcut(item):
            if sub not in embeddings_baike:
                baike_notfound.append(sub)
    if item not in embeddings_qq:
        item = item.lower()

        for sub in jieba.lcut(item):
            if sub not in embeddings_baike:
                qq_notfound.append(sub)
```
