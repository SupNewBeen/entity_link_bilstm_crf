```python
# coding:u8
import pandas as pd
import numpy as np
import copy,gc
import lightgbm as lgb
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import f1_score,classification_report
import matplotlib.pyplot as plt
from tqdm import tqdm as tqdm
import seaborn as sns
from sklearn.externals import joblib
```

```python
data = pd.read_pickle('data/step1_5.pkl')
data2 = pd.read_pickle('data/step2.pkl')
data3 = pd.read_pickle('data/step5.pkl')
```

```python
for c in data2.columns:
    if c not in data.columns:
        data[c] = data2[c]
for c in data3.columns:
    if c not in data.columns:
        data[c] = data3[c]
# for c in data4.columns:
#     if c not in data.columns:
#         data[c] = data4[c]
```

```python
exclude = ['text_id',
 'kb_id',
 'train_mention',
 'label',
 'm_id','type']
# exclude = []
temp = []
for x in data.columns:
    if x not in exclude :temp.append(x)
predictor2 = temp 
```

```python
bst = joblib.load('model_link/lgb_gbdt_1.pkl')
fold_importance_df = pd.DataFrame()
fold_importance_df["feature"] = predictor2
fold_importance_df["importance"] = bst.feature_importance()
```

```python

cols = (fold_importance_df[["feature", "importance"]]
        .groupby("feature")
        .mean()
        .sort_values(by="importance", ascending=False)[:400].index)

best_features = fold_importance_df.loc[fold_importance_df.feature.isin(cols)]
name = list(best_features.sort_values(by=['importance'])['feature'])
```

```python
len(name)
```

```python
# name = data.columns
name += exclude
```

```python
#generated by data1+data2+data3
data.loc[:,name].to_pickle('data/final.pkl')
```
